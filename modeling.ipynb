```python
df = data('iris')
df.head()
```

```python
X = df[['Sepal.Width', 'Petal.Width']]
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
kmeans.predict(X)
```

```python
df['clusters'] = kmeans.predict(X)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(131)
sns.scatterplot(data = df, x = 'Sepal.Width', y = 'Petal.Width')
plt.title('Original Data')

plt.subplot(132)
sns.scatterplot(data = df, x = 'Sepal.Width', y = 'Petal.Width', hue = 'Species')
plt.title('Actual Species')

plt.subplot(133)
sns.scatterplot(data = df, x = 'Sepal.Width', y = 'Petal.Width', hue = 'clusters')
plt.title('K-Means Clusters')
```

```python
with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(9, 6))
    pd.Series({k: KMeans(k).fit(X).inertia_ for k in range(2, 12)}).plot(marker='x')
    plt.xticks(range(2, 12))
    plt.xlabel('k')
    plt.ylabel('inertia')
    plt.title('Change in inertia as k increases')
```

```python
X = df[['Sepal.Width', 'Petal.Width', 'Petal.Length']]
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
kmeans.predict(X)
```

```python
df['new_clusters'] = kmeans.predict(X)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(131)
sns.scatterplot(data = df, x = 'Sepal.Width', y = 'Petal.Width', hue = 'Species')
plt.title('Actual Species')

plt.subplot(132)
sns.scatterplot(data = df, x = 'Sepal.Width', y = 'Petal.Width', hue = 'clusters')
plt.title('K-Means Clusters - 2 features')

plt.subplot(133)
sns.scatterplot(data = df, x = 'Sepal.Width', y = 'Petal.Width', hue = 'new_clusters')
plt.title('K-Means Clusters - 3 features')
```
